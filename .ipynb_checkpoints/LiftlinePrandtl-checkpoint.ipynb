# Simple Prandtl Lifting Line Implementation in Python

Implementation from "Aerodynamics, Aeronautics and Flight Mechanics" by Barnes W. McCormick, second edition, page 112-113.  This is supplemented by the implementation as presented on the Wikipedia page for lifting line theory:  https://en.wikipedia.org/wiki/Lifting-line_theory
Note that in some cases the implementation is a verbatim copy of the Wikipedia page, simply translated into markdown.

With the case of a 3D wing, the local angle of attack ($\alpha$) is adjusted as a result of the presence of wing vortices that are caused mainly because of the finite nature of the wing.

Therefore local lift is adjusted by the induced angle of attack $\alpha_i$.  This means the local lift coefficient at a spanwise location is calculated as:

$$C_l = C_{l0} + C_{l_\alpha} (\alpha - \alpha_i)$$

where $\alpha_i$ at a spanwise location $y_0$ is calculated as:

$$\alpha_i(y_0) = \frac{1}{4 \pi V} \int_{-b/2}^{b/2} \frac{d \Gamma(y)}{y_0 - y} $$

The boundary conditions for this problem is that we know that the lift is zero at the tips of the wing.


## Derivation

Nomenclature:
* $\Gamma$ is the circulation over the entire wing (m²/s)
* $ C_L $ is the 3D lift coefficient (for the entire wing)
* $ AR $ is the aspect ratio
* $ \alpha_\infty $ is the freestream angle of attack (rad)
* $ V_\infty $ is the freestream velocity (m/s)
* $ C_{D_i} $ is the drag coefficient for induced drag
* $ e $ is the Oswald planform efficiency factor

The following are all functions of the wings span-wise station $y$ (i.e. they can all vary along the wing)
* $ C_l $ is the 2D lift coefficient (units/m)
* $ \gamma $ is the 2D circulation at a section (m/s)
* $ c $ is the chord length of the local section
* $ \alpha_{geo} $ is the local change in angle of attack due to geometric twist of the wing
* $ \alpha_0 $ is zero-lift angle of attack of that section (depends on the airfoil geometry)
* $ C_{l_\alpha} $ is the 2D lift coefficient slope (units/m⋅rad, and depends on airfoil geometry, see Thin airfoil theory
* $ \alpha_i $ is change in angle of attack due to downwash
* $ w_i $ is the local downwash velocity

To derive the model we start with the assumption that the circulation of the wing varies as a function of the spanwise locations. The function assumed is a Fourier function. Firstly, the coordinate for the spanwise location <math>y</math> is transformed by <math> y = scos{\theta}</math>, where y is spanwise location, and s is the semi-span of the wing.


<img src="Pictures/Prandtl-lifting-line-coordinate-change.PNG">


and so the circulation is assumed to be (using Fourier series approximation):

$$ 
\begin{equation} 
\Gamma(y) = \Gamma(\theta) = \gamma = 4sV_\infty \sum_n{A_n \sin(n\theta})  \qquad
\label{eq:circulation}
\tag{1} 
\end{equation}
$$


Since the circulation of a section is related the $C_l$ by the equation:

$$ 
\begin{equation} 
C_l= \frac{2 \gamma}{V_\infty c}  \qquad
\label{eq:cl}
\tag{2} 
\end{equation}
$$

but since the coefficient of lift is a function of angle of attack:

$$ 
\begin{equation} 
C_l = C_{l_\alpha}(\alpha_\infty + \alpha_{geo} - \alpha_0 - \alpha_i)  \qquad
\label{eq:clmore}
\tag{3} 
\end{equation}
$$

hence the vortex strength at any particular spanwise station can be given by the equations:

$$ 
\begin{equation} 
\gamma = \frac{1}{2} V_\infty c C_{l_\alpha}(\alpha_\infty + \alpha_{geo} - \alpha_0 - \alpha_i)  \qquad\qquad
\label{eq:clmoremore}
\tag{4} 
\end{equation}
$$

This one equation has two unknowns: the value for $ \gamma $ and the value for $ \alpha_i $. However, the downwash is purely a function of the circulation only. So we can determine the value $ \alpha_i $ in terms of $ \Gamma(y) $, bring this term across to the left hand side of the equation and solve. The downwash at any given station is a function of the entire shed vortex system. This is determined by integrating the influence of each differential shed vortex over the span of the wing.

Differential element of circulation:

$$ 
\begin{equation} 
d\Gamma = 4 s V_\infty \sum_{n=1}^{\infty}n A_n \cos(n\theta)  \qquad
\label{eq:gammadif}
\tag{5} 
\end{equation}
$$

Differential downwash due to the differential element of circulation (acts like half an infinite vortex line):

$$ 
\begin{equation} 
dw_i = \frac{d\Gamma}{4\pi r}  \qquad
\label{eq:downwash}
\tag{6} 
\end{equation}
$$

The integral equation over the span of the wing to determine the downwash at a particular location is:

$$ 
\begin{equation} 
w_i = \int_{-s}^s \frac{1}{y-y_0} d\Gamma  \qquad
\label{eq:downwashmore}
\tag{7} 
\end{equation}
$$

After appropriate substitutions and integrations we get:


$$ 
\begin{equation} 
w_i = V_\infty \sum_{n=1}^{\infty} \frac{n A_n \sin(n \theta)}{\sin(\theta)}  \qquad
\label{eq:downwashmoremore}
\tag{8} 
\end{equation}
$$

And so the change in angle attack is determined by assuming small angles:

$$ 
\begin{equation} 
\alpha_i = \frac{w_i} {V_\infty}  \qquad
\label{eq:alphai}
\tag{9} 
\end{equation}
$$

By substituting equations $\eqref{eq:downwashmoremore}$ and $\eqref{eq:alphai}$ into RHS of equation $\eqref {eq:clmoremore}$ and equation $\eqref {eq:circulation}$ into the LHS of equation $\eqref {eq:clmoremore}$, we then get:

$$ 
\begin{equation} 
4 s V_\infty \sum_{n=1}^\infty A_n \sin(n \theta) = \frac{1}{2} V_\infty c C_{l_\alpha}\left[\alpha_\infty + \alpha_{geo} - \alpha_0 - \sum_{n=1}^{\infty} \frac{n A_n \sin(n \theta)}{\sin(\theta)} \right] \qquad
\label{eq:substitutions}
\tag{10} 
\end{equation}
$$

After rearranging, we get the series of simultaneous equations:

$$ 
\begin{equation} 
\sum_{n=1}^\infty A_n \sin(n \theta) \bigg(  \sin(\theta) + \frac{n C_{l \alpha} c}{8 s} \bigg) = \frac{C_{l \alpha} c}{8 s} \sin(\theta) (\alpha_\infty + \alpha_{geo} - \alpha_0) \qquad 
\label{eq:liftline}
\tag{11} 
\end{equation}
$$

By taking a finite number of terms, equation $\eqref{eq:liftline}$ can be expressed in matrix form and solved for coefficients A. Note the left-hand side of the equation represents each element in the matrix, and the terms on the RHS of equation $\eqref{eq:liftline}$. represent the RHS of the matrix form. Each row in the matrix form represents a different span-wise station, and each column represents a different value for n.

Appropriate choices for $\theta$ are as a linear variation between $(0 , \pi )$. Note that this range does not include the values for 0 and $\pi $, as this leads to a singular matrix, which can't be solved.


$$
\begin{bmatrix}
a_{11} & a_{12} & a_{13} & \dots & a_{1n} \\
a_{21} & a_{22} & a_{23} & \dots & a_{2n} \\
\dots  & \dots  & \dots  & \dots & \dots  \\
a_{n1} & a_{n2} & a_{n3} & \dots & a_{nn} 
\end{bmatrix}
\begin{bmatrix}
x_1 \\ x_2 \\ \dots \\ x_n 
\end{bmatrix}
=
\begin{bmatrix}
a_{1,n+1} \\ a_{2,n+1} \\ \dots \\ a_{n,n+1}
\end{bmatrix}
$$

```python
# Import all modules
import numpy as np
from math import pi, sin
import matplotlib.pyplot as plt

# Example of how to solve ax = b
a = np.array([[3,1], [1,2]])
b = np.array([9,8])
x = np.linalg.solve(a, b)
print(a)
print(b)
print(x)
```

# Matrix implementation

```python
spanstations = 20 # Number of spanwise stations for discretization of wing
nFourier = spanstations # Number of Fourier terms used to approximate the circulation distribution
# The number of terms must be equal to the span stations to have enough 
# equations for unknowns.

Vinf = 15 # [m/s] True airspeed in [m/s]

cla = 5.2 # Lift curve slope of 2D wing airfoil section
alfainf = 2*pi/180 # [rad] Angle of attack of freestream
alfazero = 0 # [rad] Angle of attack of zero lift of airfoil section
alfageo = 0 # [rad] Angle of attack of geometry
b = 10 # [m] Wing span in metres
s = b/2 # [m] semispan of wing
c = 1 # [m] Chord of wing

theta = np.linspace(0.001, 0.999*pi, spanstations)

# Example of how to solve ax = b
a = np.zeros([spanstations, nFourier])
for m in range(a.shape[0]): # rows
    for n in range(a.shape[1]): # columns
        a[m,n] = sin((n + 1)*theta[m])*(sin(theta[m]) + (n + 1)*cla*c/(8*s))

b = np.zeros([spanstations])
for m in range(b.shape[0]): # rows of column vector
    b[m] = ((cla*c)/(8*s))*sin(theta[m])*(alfainf + alfageo - alfazero)
x = np.linalg.solve(a, b)
print(a)
print(b)
print(x)
```

$$ 
\sum_{n=1}^\infty A_n \sin(n \theta) \bigg(  \sin(\theta) + \frac{n C_{l \alpha} c}{8 s} \bigg) = \frac{C_{l \alpha} c}{8 s} \sin(\theta) (\alpha_\infty + \alpha_{geo} - \alpha_0) \qquad 
$$

```python
wi = np.zeros([spanstations])

for span in range(theta.shape[0]): # rows of column vector
    for n in range(x.shape[0]): # rows of column vector
        # [m/s] Calculate spanwise downwash
        sumterm = Vinf*((n + 1)*x[n]*sin((n + 1)*theta[span]))/sin(theta[span])
        wi[span] = wi[span] + sumterm # [m/s]
        
alphaind  = wi/Vinf    
clocal = cla*(alfainf + alfageo - alfazero -alphaind)


# Now do checks.  Integrate and check that lift adds up to a sensible value
# Use check case in McCormick
# Plot clocal against span.
# Make the wing an object
# Make chord and alphageo variable

plt.plot(clocal)
```
